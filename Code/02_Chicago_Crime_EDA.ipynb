# Chicago Crime EDA

In this notebook, I acquire crime data from Google Big Query using SQL. I perform EDA by dropping appropriate columns, and producing plots that show trends in crime throughout the years. I will be using 2016 crime data as part of my train data, and 2017 data as part of my test data.

## Importing python libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
projectid = "capstone-217502"
```

## Types of crimes and their counts

In order to see which violent crimes are occuring in Chicago, we need to query the count of each type of crime, and sort them in descending order. Out of the top 25 prevelant crimes, we can select the violent types (crimes usually involving people) then query the data for each type. 

```python
crimes_and_counts = pd.read_gbq("""SELECT COUNT(*) AS count_of_crimes, primary_type
                                FROM [bigquery-public-data:chicago_crime.crime]
                                WHERE year = 2016
                                GROUP BY primary_type
                                ORDER BY count_of_crimes DESC;""", project_id = projectid)
```

## Querying crime data with SQL

Now, I can query in the data for the crimes of interest for both 2016 and 2017.

```python
crime_2016 = pd.read_gbq("""SELECT * FROM [bigquery-public-data:chicago_crime.crime] WHERE year = 2016 
                            AND primary_type = 'BATTERY'
                            OR primary_type = 'CRIM SEXUAL ASSAULT'
                            OR primary_type = 'ASSAULT'
                            OR primary_type = 'ROBBERY'
                            OR primary_type = 'INTERFERENCE WITH PUBLIC OFFICER'
                            OR primary_type = 'ARSON'
                            OR primary_type = 'HOMICIDE'
                            OR primary_type = 'KIDNAPPING'
                            OR primary_type = 'INTIMIDATION'
                            OR primary_type = 'OFFENSE INVOLVING CHILDREN'
                            OR primary_type = 'SEX OFFENSE'
                            OR primary_type = 'STALKING'""", project_id = projectid)
```

```python
crime_2017 = pd.read_gbq("""SELECT * FROM [bigquery-public-data:chicago_crime.crime] WHERE year = 2017 
                            AND primary_type = 'BATTERY'
                            OR primary_type = 'CRIM SEXUAL ASSAULT'
                            OR primary_type = 'ASSAULT'
                            OR primary_type = 'ROBBERY'
                            OR primary_type = 'INTERFERENCE WITH PUBLIC OFFICER'
                            OR primary_type = 'ARSON'
                            OR primary_type = 'HOMICIDE'
                            OR primary_type = 'KIDNAPPING'
                            OR primary_type = 'INTIMIDATION'
                            OR primary_type = 'OFFENSE INVOLVING CHILDREN'
                            OR primary_type = 'SEX OFFENSE'
                            OR primary_type = 'STALKING'""", project_id = projectid)
```

## Dropping outliers

After mapping crime incidents for both years in Tableau, I was able to see some outliers. These crime occurrences took place outside of Chicago. I can drop these outliers by reassigning the dataframes with all values except for these outliers.

```python
crime_2016 = crime_2016[crime_2016['latitude'] != 41.697]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.626]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.519]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.426]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.409]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.134]
crime_2016 = crime_2016[crime_2016['latitude'] != 41.025]
crime_2016 = crime_2016[crime_2016['latitude'] != 40.852]
crime_2016 = crime_2016[crime_2016['latitude'] != 39.205]
crime_2016 = crime_2016[crime_2016['latitude'] != 39.204]
crime_2016 = crime_2016[crime_2016['latitude'] != 36.619]
```

```python
crime_2017 = crime_2017[crime_2017['latitude'] != 41.626]
crime_2017 = crime_2017[crime_2017['latitude'] != 41.482]
crime_2017 = crime_2017[crime_2017['latitude'] != 41.426]
crime_2017 = crime_2017[crime_2017['latitude'] != 41.308]
crime_2017 = crime_2017[crime_2017['latitude'] != 41.134]
crime_2017 = crime_2017[crime_2017['latitude'] != 41.025]
crime_2017 = crime_2017[crime_2017['latitude'] != 39.205]
crime_2017 = crime_2017[crime_2017['latitude'] != 39.204]
crime_2017 = crime_2017[crime_2017['latitude'] != 36.619]
```

## Dropping columns

```python
for df in [crime_2016, crime_2017]:
    df.drop(['unique_key', 'case_number', 'description', 'fbi_code', 'x_coordinate', 'y_coordinate', 
                 'updated_on', 'iucr', 'year', 'district', 'ward', 'community_area', 'latitude', 
                 'longitude', 'location_description', 'block', 'location'], axis = 1, inplace = True)
    df[['arrest', 'domestic']] = df[['arrest', 'domestic']].astype('int64')
```

```python
print('There are',crime_2016.isnull().sum().sum(),'null values for 2016 crime data.')
print('There are',crime_2017.isnull().sum().sum(),'null values for 2017 crime data.')
```

I dropped all irrelevant and redundant columns, and assigned the boolean columns as integers so that their values would be either 0 or 1 rather than True or False. I have eliminated all null values by dropping all columns that contained nulls.

## Making month, day, hour columns

By setting the date as the index, we can make columns involving time using the index's attributes. Having columns for month, day, and hour allows us to have cleaner plots, and will make it easier to merge the weather and crime datasets by these columns. I originally dropped year, and am recreating it below for the purpose of renaming the column the same as the weather data feature "a_year." I will also do this for month, day, and hour, and will then sort the data by the index (date).

```python
for df in [crime_2016, crime_2017]:
    df['date'] = pd.to_datetime(df['date'])
    df.set_index(df['date'],inplace=True)
    df.drop('date', 1, inplace = True)
    dates = df.index
    df['a_mo'] = df.index.map(lambda dates: dates.month)
    df['a_year'] = df.index.map(lambda dates: dates.year)
    df['a_da'] = df.index.map(lambda dates: dates.day)
    df['a_hour'] = df.index.map(lambda dates: dates.hour)
    df.sort_index(inplace=True)
```

## Incident rates for each year

I want to evaluate the crime incident rate for each year. I attribute the SQL code below to Paul Mooney, who presented this code on Kaggle. You can see the original code [here](https://www.kaggle.com/paultimothymooney/how-to-query-the-chicago-crime-dataset). I, however, will adjust the code so that I could acquire the count of crimes for each year rather than the months that have the most cases of a crime, which was Paul's intention.

```python
battery_incidents = pd.read_gbq("""SELECT
  year,
  incidents
FROM (
  SELECT
    year as year,
    COUNT(1) AS incidents,
    RANK() OVER (PARTITION BY year ORDER BY incidents DESC) AS ranking
  FROM
    [bigquery-public-data.chicago_crime.crime]
  WHERE
    primary_type = 'BATTERY'
    AND year <= 2017
  GROUP BY
    year)
WHERE
  ranking = 1
ORDER BY
  year DESC
        """, project_id = projectid)
```

```python
battery_incidents.head()
```

```python
assault_incidents = pd.read_gbq("""SELECT
  year,
  incidents
FROM (
  SELECT
    year as year,
    COUNT(1) AS incidents,
    RANK() OVER (PARTITION BY year ORDER BY incidents DESC) AS ranking
  FROM
    [bigquery-public-data.chicago_crime.crime]
  WHERE
    primary_type = 'ASSAULT'
    AND year <= 2017
  GROUP BY
    year)
WHERE
  ranking = 1
ORDER BY
  year DESC
        """, project_id = projectid)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
plt.plot(battery_incidents['year'], battery_incidents['incidents'], marker = 'o', c = 'b', label = 'Battery')
plt.plot(assault_incidents['year'], assault_incidents['incidents'], marker = 'o', c = 'r', label = 'Assault')
sns.despine()
ax.set_xlabel('Year',fontsize=19)
ax.set_ylabel('Incident Rate',fontsize=19)
plt.xticks(ha='right', fontsize=16)
plt.yticks(fontsize = 16)
plt.title('Incidents per Year', fontsize=22)
plt.legend(fontsize = 13)
plt.tight_layout();
```

These crimes have some of the highest counts, and are similar in nature (both involve either harm or threat of harm). We can see that there is a general decrease in both crimes as the years progress, but when it comes to 2016, there is a slight increase in crime count. It then stays stationary in 2017. From this plot, we can say that there is a recent increase in these crimes.

```python
sexual_assault_incidents = pd.read_gbq("""SELECT
  year,
  incidents
FROM (
  SELECT
    year as year,
    COUNT(1) AS incidents,
    RANK() OVER (PARTITION BY year ORDER BY incidents DESC) AS ranking
  FROM
    [bigquery-public-data.chicago_crime.crime]
  WHERE
    primary_type = 'CRIM SEXUAL ASSAULT'
    AND year <= 2017
  GROUP BY
    year)
WHERE
  ranking = 1
ORDER BY
  year DESC
        """, project_id = projectid)
```

```python
sex_offense_incidents = pd.read_gbq("""SELECT
  year,
  incidents
FROM (
  SELECT
    year as year,
    COUNT(1) AS incidents,
    RANK() OVER (PARTITION BY year ORDER BY incidents DESC) AS ranking
  FROM
    [bigquery-public-data.chicago_crime.crime]
  WHERE
    primary_type = 'SEX OFFENSE'
    AND year <= 2017
  GROUP BY
    year)
WHERE
  ranking = 1
ORDER BY
  year DESC""", project_id = projectid)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
plt.plot(sexual_assault_incidents['year'], sexual_assault_incidents['incidents'], marker = 'o', c = 'b', alpha=1, label = 'Sexual Assault')
plt.plot(sex_offense_incidents['year'], sex_offense_incidents['incidents'], marker = 'o', c = 'r', label = 'Sex Offense')
sns.despine()
ax.set_xlabel('Year',fontsize=19)
ax.set_ylabel('Incident Rate',fontsize=19)
plt.xticks(ha='right', fontsize=16)
plt.yticks(fontsize = 16)
plt.title('Incidents per Year', fontsize=22)
plt.legend(fontsize = 13)
plt.tight_layout();
```

By grouping sexual assault and sex offense, I can make inferences for sexual crimes. In general, there is a dramatic decrease in instances of sex offenses as the years progress until 2017 (there is a slight increase). Sexual assault, on the other hand, shows increasing and decreasing trends year to year. In 2017, there seems to be almost the same number of sexual assault cases as the early 2000's. This suggests that sexual assault is harder to investigate as opposed to sex offenses.
